````python
import pandas as pd
import numpy as np
````

````python
energy = pd.read_excel('Energy Indicators.xls', skiprows=16, skipfooter=38)
````

````python
energy.head()
````

````python
energy.drop(energy.columns[[0, 1]], axis=1, inplace=True)
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy.drop(energy.index[0], inplace=True)
````

````python
energy.head()
````

````python
energy['Energy Supply'] = energy['Energy Supply']*1000000
energy.replace(r'\..*\.', np.nan, inplace=True, regex=True)
````

````python
energy.head()
````

````python
# Replace country names with the specified values
energy.replace(to_replace=[
    "Republic of Korea",
    "United States of America",
    "United Kingdom of Great Britain and Northern Ireland",
    "China, Hong Kong Special Administrative Region"],
    value=[
    "South Korea",
    "United States",
    "United Kingdom",
    "Hong Kong"],
    inplace=True)
````

````python
energy['Country'] = energy['Country'].str.replace(r" \(.*\)","")
energy['Country'] = energy['Country'].str.replace(r"\d*","")
energy['Country'] = energy['Country'].str.strip()
````

````python
# Select the top 15 countries by Rank from ScimEn
ScimEn_top15 = ScimEn[ScimEn['Rank'] <= 15]

GDP_columns = ['Country Name'] + [str(year) for year in range(2006, 2016)]
GDP = GDP[GDP_columns]

GDP.rename(columns={'Country Name': 'Country'}, inplace=True)

# Merge the datasets
merged_df = pd.merge(ScimEn_top15, energy, how='inner', on='Country')
merged_df = pd.merge(merged_df, GDP, how='inner', on='Country')

merged_df.set_index('Country', inplace=True)

columns = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document',
         'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'] + [str(year) for year in range(2006, 2016)]

final_df = merged_df[columns]

final_df
````

````python
ContinentDict = {
    'China': 'Asia',
    'United States': 'North America',
    'Japan': 'Asia',
    'United Kingdom': 'Europe',
    'Russian Federation': 'Europe',
    'Canada': 'North America',
    'Germany': 'Europe',
    'India': 'Asia',
    'France': 'Europe',
    'South Korea': 'Asia',
    'Italy': 'Europe',
    'Spain': 'Europe',
    'Iran': 'Asia',
    'Australia': 'Australia',
    'Brazil': 'South America'
}

final_df['Continent'] = final_df.index.map(ContinentDict)

continent_summary = final_df.groupby('Continent').agg(
    size=('Continent', 'size'),
    sum_self_citations=('Self-Citations Ratio', 'sum'),
    mean_self_citations=('Self-Citations Ratio', 'mean'),
    std_self_citations=('Self-Citations Ratio', 'std')
)

print(continent_summary)
````
